```python
# 使うライブラリ
# !pip install japanize-matplotlib

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import japanize_matplotlib
japanize_matplotlib.japanize()
```

## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
df = pd.read_csv('data/hist_data.csv')
plt.hist(df['aデータ'], bins=30, color='gold', alpha=0.5, label='aデータ')
plt.hist(df['bデータ'], bins=30,color='aqua', alpha=0.5, label='bデータ')
plt.title('aデータとbデータの温度に関する度数分布')
plt.xlabel('温度(℃)')
plt.ylabel('度数')
plt.legend()
plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
df = pd.read_csv('data/scatter_data.csv')
df_class0 = df[df['class']==0]
df_class1 = df[df['class']==1]
df_class2 = df[df['class']==2]
plt.scatter(df_class0['length (cm)'], df_class0['width (cm)'], color='red', label='class0')
plt.scatter(df_class1['length (cm)'], df_class1['width (cm)'], color='blue', label='class1')
plt.scatter(df_class2['length (cm)'], df_class2['width (cm)'], color='green', label='class2')
plt.title('クラス0 〜 2のlengthとwidthに関する散布図')
plt.xlabel('length (cm)')
plt.ylabel('width (cm)')
plt.grid(True)
plt.legend()
plt.show()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
df = pd.read_csv('data/line_data.csv')
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(6, 8))
ax1.plot(df['期間'], df['件数'], marker='o', color='dodgerblue', linestyle='-')
ax1.set_title('期間ごとの件数の推移')
ax1.set_xlabel('期間')
ax1.set_ylabel('件数')
ax1.grid(True)
ax1.tick_params(axis='x', rotation=45)
ax2.bar(df['期間'], df['費用'], color=['mediumblue']*len(df['期間']), alpha=0.5)
ax2.set_title('期間ごとの費用の推移')
ax2.set_xlabel('期間')
ax2.set_ylabel('費用(円)')
ax2.grid(True)
ax2.tick_params(axis='x', rotation=45)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
